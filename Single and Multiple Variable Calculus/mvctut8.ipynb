```python
#Step1:Solution of IVP
from sympy import *
t = symbols("t")
Q = Function("Q")(t)
L,R,C,E=2,24,0.005,12
DE=Eq(L*Q.diff(t,t)+R*Q.diff(t)+(1/C)*Q,E)
display(DE)
GS=dsolve(DE,Q)
display("General Sol",GS)
a,b,A,B=0,0.001,0,0
PS=dsolve(DE,Q,ics={Q.subs(t,a):b, Q.diff(t).subs(t,A):B})
display("Particular Sol",PS)
#Step2:Finding Current and plot for Q and I
Q=(-0.04425*sin(8*t)-0.059*cos(8*t))*exp(-6*t)+0.006
display("Charge=",Q)
I=Q.diff(t)
display("Current=",I)
Pt=plot(Q,I,(t,0,5),show=False)
Pt.show()
```
