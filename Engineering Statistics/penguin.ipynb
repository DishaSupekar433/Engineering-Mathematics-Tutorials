```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn import metrics
training = pd.read_csv('/content/penguins_size - penguins_size.csv')
training
```

```python
x=training.drop('species',axis=1)
x
```

```python
y=training['species']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.66, random_state =100)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
pred = model.predict(X_train)
print(metrics.classification_report(y_train, pred))
print(metrics.confusion_matrix(y_train, pred))
```
