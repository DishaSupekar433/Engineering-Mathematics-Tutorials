```python
import pandas as pd
data = pd.read_csv('/content/TSLA.csv')
data
```

```python
from numpy import *
from sklearn.linear_model import LinearRegression
high = array(data['High']).reshape(-1,1)
low = array(data['Low'])
model=LinearRegression()
model.fit(high,low)
result=model.predict(high)
print('Intercept:', model.intercept_)
print('Slope:', model.coef_)
print("Original : ",low[1002:1012])
print("Predicted : ",result[1002:1012])
```

```python
import pandas as pd
import statsmodels.api as sm
data =  pd.read_csv('/content/TSLA.csv')
df = pd.DataFrame(data) 
x = df[['High']]
y = df['Low']
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
print_model = model.summary()
print(print_model)
```
