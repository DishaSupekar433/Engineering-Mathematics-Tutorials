```python
from sympy import*
x,y,z=symbols('x,y,z')
print('finding first and second order partial derivative of multivariable function ')
f=sqrt(9-x**2-y**2)
p=diff(f,x)
q=diff(f,y)
r=diff(p,x)
s=diff(p,y)
t=diff(q,y)
f1=f.subs([(x,1),(y,2)])
p1=p.subs([(x,1),(y,2)])
q1=q.subs([(x,1),(y,2)])
r1=r.subs([(x,1),(y,2)])
s1=s.subs([(x,1),(y,2)])
t1=t.subs([(x,1),(y,2)])
display('f=',f, 'fx=',p, 'fy=',q, 'fxx=',r, 'fxy=',s, 'fyy=',t)
display('f=',f1, 'fx=',p1, 'fy=',q1, 'fxx=',r1, 'fxy=',s1, 'fyy=',t1)
```
