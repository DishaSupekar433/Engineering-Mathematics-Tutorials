# **Evaluation of double integral when limits are given**

Q1.Evaluation of double integrals in Cartesian form when all the four limits of integration are
constants.

```python
from sympy import*
x,y=symbols('x,y')
x1,x2,y1,y2=1,2,0,3
f=x**2*y
display(f)
DI=integrate(f,(x,x1,x2),(y,y1,y2))
display(DI)
```

```python
#detailed 

from sympy import*
x,y=symbols('x,y')
x1,x2,y1,y2=0,1,0,1-x
f=x+y
display(f)
#DI=integrate(f,(y,y1,y2),(x,x1,x2))
#display(DI)
I=integrate(f,y)
print('\n expreesion for inermost integral')
display(I)
VI=I.subs(y,y2)-I.subs(y,y1)
print('\n value of inermost integral within (y1,y2)')
display(VI)
DI=integrate(VI,x)
print('\n expression for outermost integral')
display(DI)
VDI=DI.subs(x,x2)-DI.subs(x,x1)
print('\n value of double integral within (x1,x2)')
display(VDI)
```
