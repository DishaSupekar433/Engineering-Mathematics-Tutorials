Tutorial 7: Find an area of the region

Type 1 Problems:

```python
#Ex.1 Find the area bounded by curve y = x**2 and line y = x

from sympy import *
x,y = symbols('x y')
f=1
y1=x
y2=x*x
Pt=plot(y1, y2, (x,-1,2),title="area bounded by curve")
P,Q=solve((y-y1, y-y2), x,y)
print("P:",P, "Q:",Q)
x1=0
x2=1
print("consider the strip parallel to Y-axis")
I=integrate(f,(y, y2,y1), (x,x1,x2))
display("Area of the region is:",I)
```
