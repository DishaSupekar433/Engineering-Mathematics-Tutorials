```python
import numpy as np
import numpy.linalg as la
from numpy.linalg import matrix_power as mpow
A = np.array([[2,1,-3],[-3,-6,1],[1,1,1]])
print("A:\n",A)
product=A@A
print("A^2:\n",product)
power=mpow(A,30)
print("A^30:\n",power)
det_A=la.det(A)
print("determinant of A:",det_A)
inv_A=la.inv(A)
print("inverse of A:",det_A)
rank=np.linalg.matrix_rank(A)
print("rank of matrix A:",rank)
trace=np.trace(A)
print("trace:",trace)
```
