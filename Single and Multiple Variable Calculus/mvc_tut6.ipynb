a. Evaluate ∫ ∫ ∫ (x-2y+z)dzdydx , z=(0,x+y), y=(0,x**2), x=(0,1)

```python
import sympy as sp
x,y,z=sp.symbols('x,y,z')
I=sp.integrate(x-2*y+z,(z,0,x+y),(y,0,x**2),(x,0,1))
print(I)
```

b. Evaluate ∫ ∫ ∫ ysin(x)+zcos(x) over the region 0 <= x <= pi, 0 <= y <= 1 and -1 <= z <= 1, x=(0,pi), y=(0,1), z=(-1,1)

```python
import sympy as sp
x,y,z=sp.symbols('x,y,z')
import math
p=math.pi
I=sp.integrate(y*sp.sin(x)+z*sp.cos(x),(z,-1,1),(y,0,1),(x,0,p))
print(I)
```

c. Evaluate ∫ ∫ ∫ (x+y+z) dx dy dz, y=(x-z,x+z), x=(0,z), z=(-1,1)
