```python
import sympy
t=sympy.symbols('t')
a=sympy.symbols('a',real=True, positive=True)
b=sympy.symbols('b',real=True, positive=True)
exp=sympy.exp
sin=sympy.sin
cos=sympy.cos
f=[1,  
   t,
   exp(a*t),
   exp(-a*t),
   sin(a*t),
   cos(a*t),
   t*exp(-a*t),
   t*exp(-a*t),
   t**2exp(a*t),
   exp(a*t)*sin(b*t),
   exp(a*t)*cos(b*t),
   exp(-a*t)*sin(b*t),
   exp(-a*t)*cos(b*t)
   ]
print("f = ",f)
```
