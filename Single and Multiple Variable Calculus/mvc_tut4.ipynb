3D Plot with Level Curves (Contour Map)

```python
from numpy import *
from matplotlib.pyplot import *
from mpl_toolkits import mplot3d
x=outer(linspace(-10,10,30),ones(30))
y=2*x.copy().T
fig=figure()

ax=axes(projection='3d')
z=(9-x**2-y**2)
ax.plot_surface(x,y,z,cmap='viridis',edgecolor='lavender')
show()
```

```python
from numpy import *
from matplotlib.pyplot import *
from mpl_toolkits import mplot3d
x=outer(linspace(-10,10,30),ones(30))
y=x.copy().T
x,y=meshgrid(x,y)
fig=figure()
ax=axes(projection='3d')
z=log(x**2+y**2)
ax.plot_surface(x,y,z,cmap='viridis',edgecolor='green')
ax.set_title('surface plot')
show()
levels=array([5,-5,10,12])
xlabel('x-axis')
ylabel('y-axis')
title('level curves of the function z=f(x,y)')
co=contour(x,y,z)
```

Maxima Minima

```python
from sympy import *
x,y=symbols('x,y')
f=x**3+y**3-63*(x+y)+12*x*y
p=diff(f,x) ; q=diff(f,y)
display(p,q)
A,B,C,D=nonlinsolve([p,q],(x,y))
print(A,B,C,D)
r=diff(f,x,x)
s=diff(f,x,y)
t=diff(f,y,y)
d=r*t-s**2
display(r,s,t,d)
print('to check nature of stationary points')
d1=d.subs([(x,A[0]),(y,A[1])])
display(d1) 
r1=r.subs([(x,A[0]),(y,A[1])])
display(r1)   
fmax=f.subs([(x,-7),(y,-7)])    
display(fmax)   
if(d1<0):
  print("no maximum/no minimum at A")
if(d1>0 and r1>0):
  print("f attains minimum at A")
if(d1>0 and r1<0):
  print("f attains maximum at A")
d2=d.subs([(x,B[0]),(y,B[1])])
display(d2) 
r2=r.subs([(x,B[0]),(y,B[1])])
display(r2)     
if(d2<0):
  print("no maximum/no minimum at B")
if(d2>0 and r2>0):
  print("f attains minimum at B")
if(d2>0 and r2<0):
  print("f attains maximum at B")
d3=d.subs([(x,C[0]),(y,C[1])])
display(d3) 
r3=r.subs([(x,C[0]),(y,C[1])])
display(r3)     
if(d3<0):
  print("no maximum/no minimum at C")
if(d3>0 and r3>0):
  print("f attains minimum at C")
if(d3>0 and r3<0):
  print("f attains maximum at C")
d4=d.subs([(x,D[0]),(y,D[1])])
display(d4) 
r4=r.subs([(x,D[0]),(y,D[1])])
display(r4)     
if(d4<0):
  print("no maximum/no minimum at D")
if(d4>0 and r4>0):
  print("f attains minimum at D")
if(d4>0 and r4<0):
  print("f attains maximum at D")
```
