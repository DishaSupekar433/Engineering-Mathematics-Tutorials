```python
import numpy as np
A = np.array([[2,1,-3],[-3,-6,1],[1,1,1]])
B=A@A
print("A=",A)
print("Square of matrix A is:")
print("A^2=",B)
C=A@A@A
print("Cube of matrix A is:")
print("A^3=",C)
D=A@A@A@A@A
print("Fifth power of matrix A is:")
print("A^5=",D)
E=A@A@A@A@A@A@A@A@A@A
print("Tenth power of matrix A is:")
print("A^10=",E)
```

```python
import numpy as np
from numpy.linalg import matrix_power as mpow
A=np.array([[1,2,3],[3,3,4],[4,5,6]])
print("The given matrix:")
print("A=\n",A)
B=mpow(A,2)
print("Second power of A:")
print("A^2= \n",B)
C=mpow(A,3)
print("Third power of A:")
print("A^3= \n",C)
D=mpow(A,4)
print("Fourth power of A:")
print("A^4=\n",D)
X=mpow(A,25)
print("Twentyfifth power of A:")
print("A^25=\n",X)
M=mpow(A,100)
print("Hundredth power of A:")
print("A^100=\n",M)
N=mpow(A,500)
print("Five hundredth power of A:")
print("A^500=\n",N)
K=mpow(A,100000)
print("One lakh power of A:")
print("A^100000=\n",K)
```
