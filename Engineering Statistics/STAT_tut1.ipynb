```python
import statistics as st

L = [35,49,225,50,30,65,40,55,52,76,48,325,47,32,60]
print("Dataset : ",L)
L.sort()
print("Sorted dataset : ",L)
med = st.median(L)
print("Median : ",med)
mean = st.mean(L)
print("Mean : ",mean)
```

```python
import pandas as pd
data = [35,49,225,50,30,65,40,55,52,76,48,325,47,32,60]
df = pd.DataFrame(data)
med = df.median()
print("Median : ",med)
mean = df.mean()
print("Mean : ",mean)
```

```python
import statistics as st

L = [20,19,17,18,17,17,17,17,17,18,18,19,17,19,18,17,19,18,19,17]
print("Dataset : ",L)
L.sort()
print("Sorted dataset : ",L)
med = st.median(L)
print("Median : ",med)
mean = st.mean(L)
print("Mean : ",mean)
mode = st.mode(L)
print("Mode : ",mode)
```
