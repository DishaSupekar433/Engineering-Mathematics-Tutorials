```python
#1
from sympy import*
x,D=symbols('x,D')
y=Function('y')(x)
a,b,g=1,4,exp(3*x)

DE=Eq(a*y.diff(x,x)+b*y.diff(x),g)
display(DE)

AE=Eq(a*D**2+b*D,0)
display(AE)

roots=solve((a*D**2+b*D,0))
display(roots)
dsolve(DE,y)
```
